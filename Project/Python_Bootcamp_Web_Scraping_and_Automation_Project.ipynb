```python
# Importing Libraries
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup
```

```python
# Parameters Decleration
NAME = []
AGE = []
OVA = []
POT = []
TEAM_NAME = []
CONTRACT = []
VALUE = []
WAGE = []
TOTAL = []
```

```python
# Fetching Data from Website 
a = [0, 60, 120, 180, 240, 300, 360, 420, 480, 540]
data = []
for i in a:
  url = f'https://sofifa.com/?offset={i}'
  response = requests.get(url)
  soup = BeautifulSoup(response.content, 'lxml')
  for i in soup.findAll('td'):
    data.append(str(i))
  for i in range(0,540):
    temp = (re.sub('<a class="tooltip".*-ellipsis">|<td class=.*">|€|<td class="col-name">|<img alt=.*/></figure>|<div class=.*transparent">|<img alt="".*png"/>|<a href=.*/">|</a><div class="sub">|</figure>|</td>|</span>|</div>|</a><img alt="".*</a>|<div.*|<figure.*|<span class.*On Loan','',data[i])).strip()
    if(i%9 == 1):
      NAME.append(temp)
    elif(i%9 == 2):
      AGE.append(temp)
    elif(i%9 == 3):
      OVA.append(temp)
    elif(i%9 == 4):
      POT.append(temp)
    elif(i%9 == 5):
      flag = temp.split('\n')
      TEAM_NAME.append(flag[0])
      CONTRACT.append(flag[1])
    elif(i%9 == 6):
      temp = temp.replace('M','')
      if('K' in temp):
        temp = temp.replace('K','')
        temp = str(float(temp)/1000)
      VALUE.append(temp)
    elif(i%9 == 7):
      if('K' in temp):
        temp = temp.replace('K','')
      else:
        temp = str(float(temp)/1000)
      WAGE.append(temp)
    elif(i%9 == 8):
      TOTAL.append(temp)
```

```python
df = pd.DataFrame({"Name" : NAME,
                   "Age" : AGE,
                   "Overall_Rating" : OVA,
                   "Potential" : POT,
                   "Team_Name" : TEAM_NAME,
                   "Contract_Year" : CONTRACT,
                   "Value(M)" : VALUE,
                   "Wage(K)" : WAGE,
                   "Total_Stats" : TOTAL})
df
```

```python
df.to_csv('fifa.csv')
```
